**1.Cargue en un Dataframe los datos de los accidentes de los últimos 3 años**

Los archivos se leen de manera individual y posteriormente se combinan en un único dataframe.

```python

df2020 = pd.read_csv("../input/atus_anual_2020.csv", index_col=False)
df2021 = pd.read_csv("../input/atus_anual_2021.csv", index_col=False)
df2022 = pd.read_csv("../input/atus_anual_2022.csv", index_col=False)
dfAccidentes = pd.concat([df2020, df2021, df2022])
dfAccidentes
```

```python
dfAccidentes = dfAccidentes.reset_index(drop=True)
dfAccidentes
```

Verificando las columnas.

**2.Agregue una columna tipo date que englobe las columnas correspondientes a Año, Mes, Dia, Hora y Minuto.**

Se crea una nueva columna de tipo datetime utilizando la función to_datetime de pandas, especificando todos los parámetros que componen la fecha.

```python
dfAccidentes['FECHA'] = pd.to_datetime({'year': dfAccidentes["ANIO"],
                                       'month': dfAccidentes["MES"],
                                       'day': dfAccidentes["ID_DIA"],
                                       'hours': dfAccidentes["ID_HORA"],
                                       'minutes': dfAccidentes["ID_MINUTO"]})
dfAccidentes['FECHA']
```

**3.Agregue una columna para Área cuyos valores serán Urbana o Suburbana dependiendo del área donde ocurrió el accidente.**

Se emplea una función llamada fill_area para determinar si el área del accidente fue urbana o suburbana.

```python
def fill_area(row):
    if 'Sin accidente en esta zona' not in row["URBANA"]:
        return 'Urbana'
    elif 'Sin accidente en esta zona' not in row["SUBURBANA"]:
        return 'Suburbana'
    else:
        return 'NA'
print("Se ejecutó correctamente")
```

Después de definir la función fill_area, se aplica esta función a la nueva columna para completar sus datos.

```python
dfAccidentes['AREA'] = dfAccidentes.apply(lambda dato: fill_area(dato), axis=1)
print(dfAccidentes['AREA'])
```

**4.Agregue una columna Zona cuyos valores serán la zona donde ocurrió el accidente**

para agregar una nueva columna de ZONA podemos utilizar una funcion similar a la anterior con la unica diferencia de que retorne los valores de dichas columnas y no una cadena predeterminada

```python
def fill_zone(row):
    if 'Sin accidente en esta zona' not in row['URBANA']:
        return row['URBANA']
    elif 'Sin accidente en esta zona' not in row['SUBURBANA']:
        return row['SUBURBANA']
    else:
        return 'NA'
print("Se ejecutó correctamente")
```

Aplicamos la función 'fill_zone' a la nueva columna

```python
dfAccidentes['ZONA'] = dfAccidentes.apply(fill_zone, axis=1)
print(dfAccidentes['ZONA'])
```

**9.¿Cómo se compara el número de accidentes de cada año contra el anterior? ¿En qué porcentaje varió?**

```python
año_2020 = len(df2020)
año_2021 = len(df2021)
año_2022 = len(df2022)

accidentes_año = dfAccidentes["ANIO"].value_counts().reset_index(name='ACCIDENTES')
accidentes_año
```

Con esta función obtenemos el procentaje de variación

```python
def variation(row):
    if row.name < len(accidentes_año) - 1:
        año_act = row["ACCIDENTES"]
        año_ant = accidentes_año.at[row.name + 1, "ACCIDENTES"]
        variation = ((año_act-año_ant)/año_ant)*100
        return variation
    else: 
        return None

print("Terminó de ejecutarse")
```

Podemos observar que el porcentaje de variación de un año con respecto al anterior es positivo y se sitúa alrededor del 10 por ciento. Además, notamos que en los años menos recientes hay un mayor crecimiento con respecto al año anterior, lo que sugiere que el crecimiento tiende a disminuir con el tiempo.

```python
accidentes_año['VARIACION'] = accidentes_año.apply(variation, axis=1)
accidentes_año
```
